```python
def getAmazonDataFrame(filename):
  p = tf.keras.utils.get_file(filename, 
                            origin="https://s3.amazonaws.com/amazon-reviews-pds/tsv/" + filename, 
                            extract=True, cache_dir=".", cache_subdir="")
  #print(p)
  #gunzip(p)
  if os.path.exists(p[0:-3]) == False:
    with open(p[0:-3], 'wb') as f_out, gzip.open(p, 'rb') as f_in:
      shutil.copyfileobj(f_in, f_out)
    
  frame = pd.read_csv(p[0:-3], sep='\t', error_bad_lines=False)
  return frame
```

### english reviews in german dataset

```python
!pip install langdetect
from langdetect import detect
import numpy as np
import pandas
```

```python
x = getAmazonDataFrame("amazon_reviews_multilingual_DE_v1_00.tsv.gz")
```

```python
l = []
for i,row in x.iterrows():
  #print(i)
  try:
    lang = detect(row.review_body)
  except:
    pass
  l.append(lang)
  #print(i)
```

```python
unique, counts = np.unique(l, return_counts=True)
d = dict(zip(unique, counts))
print(d)
```

```python
for key, value in sorted(d.items(), key=lambda item: item[1]):
    print("%s: %s" % (key, value))
```
