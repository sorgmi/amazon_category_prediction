## create dataset

```python
import tensorflow as tf
import os

import shutil, gzip
import pandas as pd


def getCategory2IndexDict():
    d = {'Video DVD':                   0,
         'Music':                       1,
         'Books':                       2,
         'Mobile_Apps':                 3,
         'Digital_Video_Download':      4,
         'Digital_Music_Purchase':      5,
         'Toys':                        6,
         'Digital_Ebook_Purchase':      7,
         'PC':                          8,
         'Camera':                      9,
         'Wireless':                    10,
         'Electronics':                 11,
         'Video':                       12,
         'Sports':                      13,
         'Video Games':                 14,
         'Watches':                     15,
         'Shoes':                       16,
         'Home':                        17,
         'Musical Instruments':         18,
         'Baby':                        19,
         'Home Improvement':            20,
         'Home Entertainment':          21,
         'Office Products':             22,
         'Personal_Care_Appliances':    23,
         'Automotive':                  24,
         'Lawn and Garden':             25,
         'Luggage':                     26,
         'Kitchen':                     27,
         'Furniture':                   28,
         'Health & Personal Care':      29,
         'Software':                    30,
         'Grocery':                     31,
         'Pet Products':                32,
         'Beauty':                      33,
        }
    return d


def getAmazonDataFrame(filename):
    p = tf.keras.utils.get_file(filename,
                                origin="https://s3.amazonaws.com/amazon-reviews-pds/tsv/" + filename,
                                extract=False, cache_dir=".", cache_subdir="cache")
    # print(p)
    outDir = p[0:-3]

    if os.path.exists(outDir) == False:
        with open(outDir, 'wb') as f_out, gzip.open(p, 'rb') as f_in:
            shutil.copyfileobj(f_in, f_out)
    else:
        print(outDir, "already exists. Skipping unzipping")

    frame = pd.read_csv(p[0:-3], sep='\t', error_bad_lines=False)
    return frame


def getData(countryCode):  # DE, UK, US
    frame = getAmazonDataFrame("amazon_reviews_multilingual_" + countryCode + "_v1_00.tsv.gz")
    return buildDataset(frame)
```

```python
data_de = getData("DE")
dataset = buildDataset(data_de)
```

```python
def buildDataset(data, buffer=None, batchsize=128):
    d = getCategory2IndexDict()

    # TODO: also use haedline (review_headline) ?
    x = data.review_body
    y = [d[x] for x in data.product_category.values]

    if buffer is None:
        buffer = data.shape[0]

    dataset = tf.data.Dataset.from_tensor_slices( (x, y) )
    dataset = dataset.shuffle(buffer).batch(batchsize)

    iter = dataset.make_one_shot_iterator()
    #iter = dataset.make_initializable_iterator()
    
    el = iter.get_next()

    return el
 

```

```python
with tf.Session() as sess:
  output = sess.run(dataset)
  print(output)
```
