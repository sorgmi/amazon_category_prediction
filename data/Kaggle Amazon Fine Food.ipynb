# create tensorflow dataset

```
dataset = tf.data.Dataset.from_tensor_slices( (data.Text.values, data.Score.values) )
```

```
iter = dataset.make_one_shot_iterator()
el = iter.get_next()

with tf.Session() as sess:
    for i in range(10):
      x, label = sess.run(el)
      print(x, label)
  
```

### with batching

```
dataset = tf.data.Dataset.from_tensor_slices( (data.Text.values, data.Score.values) ).batch(16)

iter = dataset.make_one_shot_iterator()
el = iter.get_next()

with tf.Session() as sess:
      x, label = sess.run(el)
      print(x.shape, label.shape)
      print(x)
      print(label)
```
